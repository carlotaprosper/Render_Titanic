```python
import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
df = pd.read_csv('titanic.csv')
cols = ['Survived', 'Pclass', 'Sex', 'Age']
df = df[cols]

df['Sex'] = df['Sex'].apply(lambda x: 1 if "female" else 0)
X = df[['Pclass', 'Sex', 'Age']]
y = df.Survived

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
rf.score(X_test, y_test)
```

```python
with open("model.pkl", "wb") as archivo:
    pickle.dump(rf, archivo) # dump para archivo y escribir
```

```python
with open("model.pkl", "rb") as archivo:
    modelo = pickle.load(archivo) # load para leer el archivo
```

```python
modelo.score(X_test, y_test)
```

```python
age = 20
sex = 1
clase = 3

survived = "Sobrevivio" if modelo.predict([[age, sex, clase]])[0] else "Palmo"

respuesta = f"""Para la persona con {age} años, sexo {sex} y clase {clase}, la prediccion es {survived}"""
print(respuesta)
```
